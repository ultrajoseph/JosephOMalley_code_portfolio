
## Exploration of Tokenization Methods
Each of these sections will explore one or more different preprocessing steps used in the tokeization process.  Each method has the potential to provide insight into the underlying free text data.


### Method 1 - Simple Count Vectorizer w/ stopwords
This first version first eliminates stopwords <i.e. "the", "but", "for"> to both conserve space in the matrix and eliminate words that dont carry much meaning.  The output is a sparse matriz that has each unique tweet on the left and each unique token <word, in this case> as each column heading -- each corresponding cell has the count of how many time that word occured in a particular tweet.

```python
cv2 = CountVectorizer(binary=False, stop_words='english') #define the transformation
cv2_chat = cv2.fit_transform(blockchain_df['newtext']) #apply the transformation

print(type(cv2_chat))
##Dimensions of the first method -- cv2
print(cv2_chat.shape)

# multiset or "bag of words"
#returns all unique tokens included in the matrix
#print(cv2.get_feature_names())

##head of the sparse matrix --tweet numbers on the left, unique tokens across the top
pd.DataFrame(cv2_chat.toarray(),columns = cv2.get_feature_names()).head()
```

```python
names = cv2.get_feature_names()   #create list of feature names
print(type(names), len(names))

count = np.sum(cv2_chat.toarray(), axis = 0) # convert list to array to add up feature counts 
count2 = count.tolist()  # convert numpy array to list


count_df = pd.DataFrame(count2, index = names, columns = ['count']) # create a dataframe from the list
sorted_count = count_df.sort_values(['count'], ascending = False)
# create a dataframe from the list
print(sorted_count.head(20))
print(sorted_count.tail(10))
```


### Method 2 - Simple Count Vectorizer w/ N-Grams
This method also uses counts and includes N-Grams (2 and 3 word phrases, in this case).  These phrases may pick up context <i.e. "Blockchain regualtion"> that would otherwise be missed by a single word tokeization process.  Additionally, I choose not to remove stopwords because a 3 word phrase may be connected with a stopword.  Matrix dimensions are much larger using this method because the number of unique combinations of words (2 and 3, without stopwords) is far greater than that of single words)

```python
# try changing a few parameters
pd.set_option('display.max_columns', None)
cv5 = CountVectorizer(binary=False, ngram_range = (2, 3)) #define the transformation
cv5_chat = cv5.fit_transform(blockchain_df['newtext']) #apply the transformation

print(type(cv5_chat))
##Dimensions of the second method -- cv5
print(cv5_chat.shape)
##head of the sparse matrix --tweet numbers on the left, unique tokens across the top
pd.DataFrame(cv5_chat.toarray(),columns = cv5.get_feature_names()).head()
```

```python
names = cv5.get_feature_names()   #create list of feature names
print(type(names), len(names))

count = np.sum(cv5_chat.toarray(), axis = 0) # convert list to array to add up feature counts 
count2 = count.tolist()  # convert numpy array to list


count_df = pd.DataFrame(count2, index = names, columns = ['count']) # create a dataframe from the list
sorted_count = count_df.sort_values(['count'], ascending = False)
# create a dataframe from the list
print(sorted_count.head(20))
print(sorted_count.tail(10))
```


### Method 3 - Count vector with min/max frequency requirement 
This method also uses counts and allows you to specify the minumum and maximum frequencies that a word can occur to be included (in percentage terms)

```python
# set max df to not include anything occuring in over 50 percent of responses, and min_df to include only those in at least 1 percent of response
cv6 = CountVectorizer(binary=False, max_df = .5, min_df= .01) #define the transformation
# only asking it to make changes based on document frequency

cv6_chat = cv6.fit_transform(blockchain_df['newtext']) #apply the transformation

print(type(cv6_chat))
print(cv6_chat.shape)
pd.DataFrame(cv6_chat.toarray(),columns = cv6.get_feature_names()).head()
```

```python
names = cv6.get_feature_names()   #create list of feature names
print(type(names), len(names))

count = np.sum(cv6_chat.toarray(), axis = 0) # convert list to array to add up feature counts 
count2 = count.tolist()  # convert numpy array to list


count_df = pd.DataFrame(count2, index = names, columns = ['count']) # create a dataframe from the list
sorted_count = count_df.sort_values(['count'], ascending = False)
# create a dataframe from the list
print(sorted_count.head(20))
print(sorted_count.tail(10))
```

### Method 4 - Tf-Idf
This method gives higher weight to words that occur many times within a small number of documents (tweets in our case).  This weight is lowest when the term occurs in virtually all documents

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# now apply weights based on the inverse document frequency
tfidf = TfidfVectorizer(use_idf=True, norm=None, stop_words='english') #define the transformation
tf_chat = tfidf.fit_transform(blockchain_df['newtext']) #apply the transformation

pd.DataFrame(tf_chat.toarray(),columns = tfidf.get_feature_names()).head()
```

#### Examine the weighted feature space

```python
names = tfidf.get_feature_names()
print(type(names), len(names))

#create list of feature names
count = np.sum(tf_chat.toarray(), axis = 0) # add up feature counts 
count2 = count.tolist()  # convert numpy array to list

count_df = pd.DataFrame(count2, index = names, columns = ['count']) # create a dataframe from the list
sorted_count = count_df.sort_values(['count'], ascending = False)
# create a dataframe from the list
print(sorted_count.head(20))
print(sorted_count.tail(10))
```

```python
# Maybe Average weight of terms
count = np.mean(tf_chat.toarray(), axis = 0) # find the average weight for each feature 
count2 = count.tolist()  # convert numpy array to list
count_df = pd.DataFrame(count2, index = names, columns = ['count']) 
# create a dataframe from the list
sorted_count = count_df.sort_values(['count'], ascending = False)
# create a dataframe from the list
print(sorted_count.head(20))
print(sorted_count.tail(10))
```

```python
# max weight of a given term
count = np.max(tf_chat.toarray(), axis = 0) # find the max weight of each feature 
count2 = count.tolist()  # convert numpy array to list
count_df = pd.DataFrame(count2, index = names, columns = ['count'])
sorted_count = count_df.sort_values(['count'], ascending = False)
# create a dataframe from the list
print(sorted_count.head(20))
print(sorted_count.tail(10))
```
